```python
df_learn = pd.read_csv('../data/train.csv')
df_learn.head()
```

```python
df_distances = pd.read_csv('../data/learn_distances.csv',index_col=0)
print(df_distances.shape)
df_distances.head()
```

```python
df = pd.read_csv("../data/features/features_part_one.csv",index_col=0)
print(df.shape)
df.head()
```

```python
df_learn.dropna(inplace=True)
y = df_learn['is_duplicate']
```

```python
df.replace([np.inf, -np.inf], np.nan,inplace=True)
df.fillna(0.0,inplace=True)
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
norm = StandardScaler()

d = set()
d.add('length')
f_tags = list(set(df.columns) - d)

df_norm = pd.DataFrame(norm.fit_transform(df[f_tags]),columns=f_tags)
df_norm['length'] = df.length
df_norm.describe()
```

```python
features_dist = ['jaccard','cosine']
df_norm_dist = df_norm.join(df_distances[features_dist])

df_dist = df.join(df_distances[features_dist])

df_norm_dist.head()
```

Кросс-валидация

```python
df_norm_dist = df_norm_dist.join(df_learn.is_duplicate)
df_dist = df_dist.join(df_learn.is_duplicate)
print(df_norm_dist.shape)
print(df_dist.shape)
```

```python
df_norm_dist.dropna(inplace=True)
df_dist.dropna(inplace=True)

print(df_norm_dist.shape)
print(df_dist.shape)
```

```python
from sklearn.model_selection import KFold,StratifiedKFold,StratifiedShuffleSplit
from sklearn.metrics import accuracy_score,confusion_matrix, log_loss
from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

rf = RandomForestClassifier()
log_r = LogisticRegression(C=10)
X = df_dist[f_tags + features_dist +['length'] ]
# X = df_dist[ f_tags]
y = df_dist.is_duplicate.values

skf = StratifiedKFold(n_splits=3,shuffle=True)

classifier = rf
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    classifier.fit_transform(X_train,y_train)
    y_pred = classifier.predict(X_test)
    print("log_loss: %f" % log_loss(y_true=y_test,y_pred=y_pred))
    print("accuracy: %f" % accuracy_score(y_true=y_test,y_pred=y_pred))
    print("cm:{}".format(pd.DataFrame(confusion_matrix(y_true=y_test,y_pred=y_pred))))
```
